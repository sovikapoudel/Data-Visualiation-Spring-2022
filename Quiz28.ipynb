#Using the workout dataset, create a scatterplot with as many features as possible. Can you get 5 or six variables represented in one graphic?

```python
import pandas as pa
import matplotlib.pyplot as plt
import seaborn as sns


df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')

df.head()
```

```python
ax = sns.scatterplot(data = df, 
                x = 'workout_type', 
                y = 'distance',
                hue = 'workout_type',
                size = 'calories',
                legend = 'brief')
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1)) #this is hard to read because of it's size so I will increase the size in next part. and try ddifferent charts.
```

```python
sns.set(rc={"figure.figsize":(35, 10)}) #width=35, height=10, I found the way to increase the size so it's easier to visualize. 
#This line of code let figure size by passing dictionary to rc parameter with key 'figure.figsize' in seaborn set method.
sns.heatmap(df.corr(), annot=True, linewidths=.5)
```

```python
sns.set(rc={"figure.figsize":(25, 8)}) #width=25, height=8, I found the way to increase the size tso that 
ax = sns.scatterplot(data = df, 
                x = 'time', 
                y = 'distance',
                hue = 'workout_type',
                size = 'time',
                x_bins = "avg_speed") #not sure what this is doing

sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
```

```python
g = sns.PairGrid(df, hue=df.groupby('workout_type'))
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
```
