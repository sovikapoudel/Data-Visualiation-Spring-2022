#Importing the necessary libraries and loading a dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
dfVax= pd.read_csv("https://raw.githubusercontent.com/sovikapoudel/Data-Visualiation-Spring-2022/main/vaccination-data-WHO.csv") #Loading the dataset and storing in pandas dataframe.
dfVax.head(5)
```

###Checking the types of data first. It is very important to check and confirm the data types a part of examining Data.

```python
dfVax.dtypes
```

## Now, Let see how many countries are there and number of total vaccination by grouping them

```python
dfVax.COUNTRY.agg('count')
```

##The dateframe aleady grouped by the countries so, I do not need to group them.

###There are 228 unique countries in this dataframe. So, lets find out if people are vaccinated or not. 

##First, there are some missing datas (NaN) as this is a timely data so I am filtering out and removing all the NaN so let's drop those tables.

```python
df2=dfVax.dropna() #just a simple dropna() function removes all the null tables in the dataframe
df2
```

##Now we know there are total of 174 unique countries without null entries in the dataframe. So, let find first if all the countries has a vaccination record or not!

###As this is a timely data, the dataframe is sorted by date so Let's find the maximum number of vaccination by each country and which vaccine were used

```python
df3 = df2.loc[df2['TOTAL_VACCINATIONS']>0,['DATE_UPDATED','COUNTRY', 'TOTAL_VACCINATIONS', 'VACCINES_USED']] # giving a condition where the number of people vaccination is greater than 0.
df3
```

###Now we know there are people from every country who are vaccinated but let's find out if they all of them are fully vaccinated or not, By that I mean how many of them have received all the doses.

```python
highest_vax = df2.loc[df2['PERSONS_FULLY_VACCINATED_PER100']>99,['DATE_UPDATED','COUNTRY', 'PERSONS_FULLY_VACCINATED_PER100']] # giving a condition where the number of people vaccination is greater than 99 which is basically 100. No country are 100% vaccinated so I gave 99% condition.
highest_vax
```

```python
df4 = df2.loc[df2['PERSONS_FULLY_VACCINATED_PER100']<99,['DATE_UPDATED','COUNTRY', 'PERSONS_FULLY_VACCINATED_PER100']]
df4.sort_values(by = "PERSONS_FULLY_VACCINATED_PER100", ascending =False)
```

# So, let's find out how many population of United States is vaccinated. My guess is somewhere around 60%

```python
USA=dfVax.loc[dfVax['COUNTRY']=='United States of America',['DATE_UPDATED','COUNTRY', 'PERSONS_FULLY_VACCINATED_PER100']] #Had to look at my csv data to find how US name was recorded!
USA
```

Now we know, there exists USA so let's sort out the countries and get only required tables.

```python
dfVax2 = dfVax.loc[dfVax['TOTAL_VACCINATIONS']>0,['DATE_UPDATED','COUNTRY', 'TOTAL_VACCINATIONS', 'TOTAL_VACCINATIONS_PER100','PERSONS_FULLY_VACCINATED_PER100']]
G7countries = ['France', 'Canada', 'United States of America', 'Germany', 'Italy', 'Japan', 'United Kingdom']
MyCountries=dfVax2[dfVax2["COUNTRY"].isin(G7countries)]
MyCountries
```

Looks like US is the last one among all the G7 countries. Now, I have all the required countries so, I am going to create a graph now!I am going to create step by step.

```python
lineplot1 = MyCountries.plot.line(x='DATE_UPDATED', y='PERSONS_FULLY_VACCINATED_PER100')
```
