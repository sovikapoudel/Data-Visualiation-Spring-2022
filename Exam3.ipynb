## (10 points) Create a barchart for most goals scored by nationality.


```python
df1=df.groupby('nationality').Goals.agg(['count'])
df1.head(10)
#this shoud give me that number of game by nationality.
```

```python
df.groupby('nationality')
```

```python
df1=df.groupby('nationality').Goals.agg(['sum'])
df1.head(10)
#sum shoud give me that most goal scored by nationality, I am compare the count and sum to see the diiference in numbers. 
#In count, it is just counting the extries of the column "Goals" where in sum it is actually adding the number of goal grouped by nationality.
```

## Now, Creating a barchart for most goals scored by nationality.


```python
dfbar = df.groupby(['nationality'])['Goals'].sum()
dfbar.plot(kind='bar', title="Barchart of Goals")
plt.show()
```
