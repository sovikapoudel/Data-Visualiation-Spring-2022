#Now for room_type I am following Dr. Jacob's method

```python
rmtyp = sorted(df.room_type.unique())
rmtyp #this prints out 3 types and that's the range
rmtypdict = {}
for i in range(3):
  rmtypdict[rmtyp[i]] = i

rmtypdict
##This will create a dictionary and assign the value to each category.
```

```python
df2.room_type.map(rmtypdict) #here I have mapped all the room_type to the dictionary I created in previous step.
```

```python
df2.room_type = df2.room_type.map(rmtypdict)

df2.head()
```
