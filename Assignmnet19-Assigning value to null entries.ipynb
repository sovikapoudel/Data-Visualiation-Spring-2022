#Assigning value to entire dataframe

```python
df2 = pa.read_csv("https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv") 
df2.fillna(value = 88888, #The fillna() method replaces the NULL values with any specified value, I am assigning 8888 in this case
          inplace = True) 
df2.head()
##I had to reassign the whole data frame to run this code as it was pulling the value from previous section
```

## inplace=True is the keyword in a pandas method that changes the default behaviour such that the entries in dataframe has no value and doesnot return anything, if inplace=true then it will overwrite the value

```python
df3 = pa.read_csv("https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv") 
df3.fillna(value = 8888, #The fillna() method replaces the NULL values with any specified value, I am assigning 8888 in this case
          inplace =True) 
df3.host_name
```

```python
df3.host_name.loc[df3['host_name']== 8888] #this is the way to check if the value has been assigned or not.
```

#Part 2-   Impute for atleast two columns that are null in an approriate fashion.

# Let's find how many null value are in host_name

```python
dfhost_name = pa.read_csv("https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv") 

dfhost_name['host_name'].isna().sum() #this is givng total number of null value in host_name column.
```

# Now, Let's get rid of thouse 21 null value.

```python
dfhost_name.fillna(value = 8888, #The fillna() method replaces the NULL values with any specified value, I am assigning 8888 in this case
          inplace =True)
dfhost_name['host_name'].isna().sum() #this should give 0 as I have already assign null host_name some value (9999)
```

## Again, let's find how many null value are in name

```python
dfname=pa.read_csv("https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv") 
dfname['name'].isna().sum()
```

#There are 16 null under name so let's replace them by 'abc', string in this case

```python
dfname.fillna(value = 'abc', #The fillna() method replaces the NULL values with any specified value, I am assigning 8888 in this case
          inplace =True)
dfname.loc[dfname['name'] == 'abc'] #this way I can check if it has assigned 'abc' value to the targeted column
```
