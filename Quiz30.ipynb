```python
import pandas as pa
import matplotlib.pyplot as plt
import seaborn as sns

df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')

df.head()
```

# For exam, I am thinking about the 
#Barchart
#Histogram
#Scatterplot
#Violinplot 

```python
ax = df.groupby('workout_type').calories.agg('mean').plot.barh()
ax.set(title = "Average calories by workout_type")
ax.set(xlabel = 'Calories')
plt.show()
```

```python
ax = sns.scatterplot(data = df, x = 'calories', y = 'distance', hue = 'time')
ax.set(title = "Scatter")

plt.show() #However this is not be a perfect example of scatterplot that I wanted for exam.
```

```python
ax = df.calories.hist(bins = 25)
ax.set(title = "Calories in Histogram")
plt.show()
```
