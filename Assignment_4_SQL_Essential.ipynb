Now Let's get the dataset.

```python
%%bigquery --project data-vis-assignment-338801
SELECT *
  From `bigquery-public-data.austin_bikeshare.bikeshare_trips`
  limit 20
   ###using limit will gives exactly how many table youwant to generate.
```

I think 20 should be enought for this part of assignment. I tried to pull all the table but it was taking a lot of time (more than 2 minutes) then I gave up. So, I will be using 20 rows.





I am going to compute the average time for a trip based on each starting point in next step.

```python

%%bigquery --project data-vis-assignment-338801
SELECT start_station_id as Starting_Point, AVG(duration_minutes) as Average_Time
From `bigquery-public-data.austin_bikeshare.bikeshare_trips`
Group BY start_station_id
```

### How many trips start at each starting point?


Now, I am going to find out how many trips started at each starting point. For starting point I am going to use start_station_id column

```python
%%bigquery --project data-vis-assignment-338801
SELECT start_station_id as Starting_Point, COUNT(start_station_id) AS Num_Of_Trips
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_id
LIMIT 20
```
