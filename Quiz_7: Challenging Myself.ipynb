###Task to do!

Examine the average trip on Austin bike share grouping by property_type column from the bikeshare_stations table. Is average time on trips starting from 'parkland' longer than from 'sidewalks'? To answer this question you must join the bikeshare_trips and bikeshare_stations on the start_station_name and the name from their respective tables.

SQL Joins

```python
%%bigquery --project data-vis-assignment-338801 
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_stations`
WHERE property_type != "NONE"
Limit 10
```

First, let's gather the data about the property type and it's count.

```python
%%bigquery --project data-vis-assignment-338801 
SELECT property_type as Type, COUNT(property_type ) as Num
FROM `bigquery-public-data.austin_bikeshare.bikeshare_stations` 
WHERE property_type != " " ##There's one blank entry in the dataset under property_type so I am using the where clause to remove that from my output
group by  property_type
LIMIT 20 #just checking if this pulls out the all the informations, we only have total of 6 entries for this query though
```

Now, we have got the total numbers of each property types. Let's join these both datasets for our next part.

```python
%%bigquery --project data-vis-assignment-338801 
SELECT trip.start_station_id, trip.start_station_name, station.property_type, trip.duration_minutes as Trip_Duration
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips` as trip INNER JOIN `bigquery-public-data.austin_bikeshare.bikeshare_stations` as station 
ON trip.start_station_name=station.name AND station.property_type != "None" AND station.property_type != " "
LIMIT 20
```

Now, Let's find out the average duration time of each property type by grouping them together by property_type

```python
%%bigquery --project data-vis-assignment-338801 
SELECT station.property_type, AVG(trip.duration_minutes) as Average_Trip_Duration
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips` as trip INNER JOIN `bigquery-public-data.austin_bikeshare.bikeshare_stations` as station 
ON trip.start_station_name=station.name AND station.property_type != "None" AND station.property_type != " "
GROUP BY station.property_type
LIMIT 20

```

Now we know the total numbers of bike, I will now examine if there's any bike thta has not been used yet.

```python
%%bigquery --project data-vis-assignment-338801
SELECT Start_station_name, bikeid as Bike_Id
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
# Where bikeid != " " 
Group by Start_station_name, bikeid
Order BY bikeid ASC
LIMIT 100
```

```python
%%bigquery --project data-vis-assignment-338801
SELECT SUM(duration_minutes) as total_duration_minutes, bikeid as Bike_Id
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
# Where bikeid != " " 
Group by  bikeid 
Order BY total_duration_minutes DESC
LIMIT 100
```
