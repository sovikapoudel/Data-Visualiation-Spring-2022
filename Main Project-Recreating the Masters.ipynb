###First, there are some missing datas (NaN) as this is a timely data so I am filtering out and removing all the NaN so let's drop those tables.

```python
df2=dfVax.dropna() #just a simple dropna() function removes all the null tables in the dataframe
df2
```

###Now we know there are total of 174 unique countries without null entries in the dataframe. So, let find first if all the countries has a vaccination record or not!

###As this is a timely data, the dataframe is sorted by date so Let's find the maximum number of vaccination by each country and which vaccine were used

```python
df3 = df2.loc[df2['TOTAL_VACCINATIONS']>0,['DATE_UPDATED','COUNTRY', 'TOTAL_VACCINATIONS', 'VACCINES_USED']] # giving a condition where the number of people vaccination is greater than 0.
df3
```

###Now we know there are people from every country who are vaccinated but let's find out if they all of them are fully vaccinated or not, By that I mean how many of them have received all the doses.

```python
highest_vax = df2.loc[df2['PERSONS_FULLY_VACCINATED_PER100']>99,['DATE_UPDATED','COUNTRY', 'PERSONS_FULLY_VACCINATED_PER100']] # giving a condition where the number of people vaccination is greater than 99 which is basically 100. No country are 100% vaccinated so I gave 99% condition.
highest_vax
```

```python
df4 = df2.loc[df2['PERSONS_FULLY_VACCINATED_PER100']<99,['DATE_UPDATED','COUNTRY', 'PERSONS_FULLY_VACCINATED_PER100']]
df4.sort_values(by = "PERSONS_FULLY_VACCINATED_PER100", ascending =False)
```

###I would have to get the raw unrefined data for the graph I wanted to recreate. The dataframe I have is already grouped by country and there is only one date for each country which is updated date. It is available in the same website I referenced on the very begining of project.

```python
import pandas as pd
G7vax= pd.read_csv("https://raw.githubusercontent.com/sovikapoudel/Data-Visualiation-Spring-2022/main/vax-csv.csv") #Loading the dataset and storing in pandas dataframe.
G7vax.head(5)
```

##Removing NANs

```python
G7vax3=G7vax.dropna()
G7vax3
```

```python
G7Vax2 = G7vax3.loc[G7vax3['total_vaccinations']>0,['date','country', 'people_fully_vaccinated_per_hundred', 'total_vaccinations_per_hundred','people_vaccinated_per_hundred']]
G7list = ['France', 'Canada', 'United States', 'Germany', 'Italy', 'Japan', 'United Kingdom']
G7df=G7vax3[G7vax3["country"].isin(G7list)]
G7df
```

###Now, I have cleaned all the NaN values and refined only G7 countries. In this dataframe there are multiple ddate and mutiple entries for same country. Now, let's clean again and create a visualization.

###Recreating Master Graph

```python
fig, ax = plt.subplots()
sns.lineplot(x = 'date', y = 'people_fully_vaccinated_per_hundred', hue = 'country', data = G7df, ax = ax)
ax.set_title("G7 Countries Vaccination ")
ax.legend(title ="Countries")
ax.set_xlabel('Date of vaccination')
ax.set_ylabel("Percentage of vaccination per 100 people")

```

###Now I am almost close to the master. And, recreating master could include some changes so that it's easier for viewer to visualize so I am going to make following changes to improve in the master

*   The date in the dataframe is not in in panda dataframe so now, it needs to be changed first in order to show the date in one of the axis.
*   The lines are small which is not eay to ready so changing the line width will be more helpful.
* Some changes in legend will make graph easier to follow. Like the size and place.





```python
G7df['date'] = pd.to_datetime(G7df['date'], format="%m/%d/%Y") #here I am changing the dateformat to panda date format
```

```python
fig, ax = plt.subplots(figsize = (20,9))
sns.lineplot(x = 'date', y = 'people_fully_vaccinated_per_hundred', hue = 'country', data = G7df, linewidth = 4) #setting the linewidth for better visualization
ax.set_title("G7 Countries Vaccination ")
ax.set_xlabel('Date of vaccination')
ax.set_ylabel("Percentage of vaccination per 100 people")
#adding some changes to legend
ax.legend(title ="Countries", loc = "upper left", prop = {"size":17}) #this is for the legend
```
